```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
#Printing the stopwords in English
print(stopwords.words('english'))
```

Data Preprocessing

```python
data = pd.read_csv('/content/drive/MyDrive/Projects/Fake News Prediction/train.csv')
```

Label represents the output data
l --> Fake News 
0 --> Real News

```python
data.head()
```

```python
#Counting the number of missing values
data.isnull().sum()
```

```python
 #Replacing the null values with empty string
data = data.fillna('')
```

```python
#Merging the author name and news title
data['content'] = data['author']+ ' '+ data['title']
```

```python
# print(data['content'])
```

```python
#Separating the data & label
X = data.drop('label' , axis = 1)
Y = data['label']
# print(X)
# print(Y)
```

Stemming :
Stemming is the process of reducing a word to its Root word
ex -- actor, actress, acting --> act(Root Word)

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]', ' ', content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
data['content']= data['content'].apply(stemming)
```

```python
# print(data['content'])
```

```python
X = data['content'].values
Y = data['label'].values
```

```python
# Converting the text values to numerical data so that computer can understand
# TF stands for Term Frequency and IDF stands for Inverse document frequency.Tf counts the number of times a particular 
#word is repeating in document and get that this value is important thus assigns a numeric value for that word.IDF counts the most repeting words and detects that 
# it is not so important and reduces its importance values which was given by TF.By this they create feature vectors which is a numeric values.

vectorizer = TfidfVectorizer()
vectorizer.fit(X)
X= vectorizer.transform(X)
# print(X)
```

Splitting the dataset to training and test data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state =2)
print(X_test[0])
```

Training the Logistic Regression model

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Evaluation

```python
#Accuracy Score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("The accuracy score of the training data : ", training_data_accuracy)
```

```python
#Accuracy Score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Test data accuracy :", test_data_accuracy)
```

Making a Predictive System

```python
X_new = X_test[0]
prediction = model.predict(X_new)
print(prediction)
if prediction== 0:
  print("The news is real")
else:
  print("The news is fake")
```
